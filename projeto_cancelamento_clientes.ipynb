```python
import pandas as pd


```

```python
# importando e visualizando a tabela

tabela = pd.read_csv("/content/cancelamentos_sample.csv")

display(tabela)
```

#**Análise das informações da tabela e identificar possíveis erros e informações inúteis**

###depois de analisar as colunas é identificado colunas que não nos ajudaram e podem nos atrapalhar podendo deixar a tabela mais pesada

```python
# Vamos utilizar o "drop" para tirar a coluna "CustomerID" que não terá quaçquer utilidade em nossas análises

tabela = tabela.drop(columns="CustomerID")

display(tabela)
```

```python
# Visualizando a base de dados contendo informações sobre valores nulos

display(tabela.info())
```

```python
# Tratando valores nulos excluindo essas linhas com valores nulos

tabela = tabela.dropna()

display(tabela.info())

```

#**Analisando os dados**

```python
# Quantas pessoas cancelaram e quantas pessoas não cancelaram

display(tabela["cancelou"].value_counts())
```

```python
# Visualizando em porcentagem

display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Análise das causas do cancelamento utilizando gráficos

import plotly.express as px

for coluna in tabela.columns:
  grafico = px.histogram(tabela, x=coluna, color="cancelou")
  grafico.show()
```

# **Conclusões das análises dos gráficos**

### Todos os clientes do contrato mensal cancelam a assinatura.
### Clientes que ligam mais de 4 vezes para o Callcenter, cancelam a assinatura.
### Clientes que atrasaram mais de 20 dias, cancelam a assinatura.

# **Resoluções dos problemas**

### Oferecer descontos nos anos anuais e trimestrais.
### Criar um método que resolva o probelma do cliente em no máximo 3 ligações.
### Reformular a política para resolver atrasos em até 10 dias.




```python
# Fazendo uma filtragem para saber a porcentagem de cancelamento após as resoluções dos problemas

tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
tabela = tabela[tabela["ligacoes_callcenter"]<=4]
tabela = tabela[tabela["dias_atraso"]<=20]

display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

# **Conclusões Finais**

### Depois das possíveis resoluções utilizando filtragens e verificar a porcentagens de cancelamentos caso essas resoluções fosssem feitas, podemos observar que há uma diminuição de cancelamentos muito brusca.
